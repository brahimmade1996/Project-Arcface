```python
import os
from imutils import paths
import random
import shutil
from tqdm import tqdm
```

```python
old_path  = os.path.join(os.getcwd(), 'data/k-face/old_path').replace('\\','/')
old_path
```

```python
new_path =os.path.join(os.getcwd(), 'data/k-face/old_path').replace('old','new').replace('\\','/')
try:
    os.mkdir(new_path)
    os.mkdir(os.path.join(new_path, 'train'))
    os.mkdir(os.path.join(new_path, 'test'))

except Exception as err:
    print(err)
    
```

```python
# 폴더 내 모든 이미지 파일 가져오기


Ls= ['L1', 'L2', 'L3']
Ss=['S001', 'S002', 'S005']
Cs = ['C'+ str(i)+'.jpg' for i in range(5, 11)]
Es = ['E1', 'E2', 'E3']
```

```python
use_paths = {}
use_values = []
for root, dirs, files in os.walk(old_path):
    values = []
    path_list = root.replace('\\','/').split('/')
    if len(path_list) >=13:
        LL = path_list[11] in Ls
        SS = path_list[10] in Ss
        EE = path_list[12] in Es
        if LL and SS:
            use_values.append(root)
            use_paths[path_list[9]]=use_values

use_paths
        
```

```python
train_path = 'C:/Users/shiney/Documents/GitHub/Project-Arcface/data/k-face/new_path/train'
test_path = 'C:/Users/shiney/Documents/GitHub/Project-Arcface/data/k-face/new_path/test'

for people_name in use_paths:
    dirs_path = use_paths[people_name]
    try:
        os.mkdir(os.path.join(train_path, people_name))
        os.mkdir(os.path.join(test_path, people_name))
    except Exception as err:
        print(err)
    
    # 사람 안에 폴더마다 들어가기
    new_img_paths=[]
    for dir_path in dirs_path:
        # 폴더 안에 있는 이미지 list
        img_paths = list(paths.list_images(dir_path))
        
        img_paths = [img_path for img_path in img_paths if img_path.replace('\\','/').split('/')[-1] in Cs]
        #print(img_paths, len(img_paths))
        new_img_paths.append(img_paths)
    new_img_paths= sum(new_img_paths, [])
    random.shuffle(new_img_paths)

    for i,img_path in enumerate(new_img_paths):
        print(i, img_path)
        img_name = img_path.replace('\\','/').split('/')
        print(img_name)

        img_name = '_'.join(img_name[10:])
        print(img_name)
        if i < 30:
            new_img_path = f'{test_path}/{people_name}/{img_name}'
        else:
            new_img_path = f'{train_path}/{people_name}/{img_name}'
        print(new_img_path)
        shutil.copy(img_path, new_img_path)            
```
